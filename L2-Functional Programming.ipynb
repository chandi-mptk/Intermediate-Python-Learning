**Functional Programming**

```python
def apply_twice(func, arg):
    return func(func(arg))

def add_five(x):
    return x + 5

print(apply_twice(add_five, 10))
```

**Pure Functions**
Always return same values for same argument


```python
def pure_function(x, y):
  temp = x + 2*y
  return temp / (2*x + y)
```

**Lambdas**

```python
def my_func(f, arg):
  return f(arg)

my_func(lambda x: 2*x*x, 5)
```

```python
#named function
def polynomial(x):
    return x**2 + 5*x + 4
print(polynomial(-4))

#lambda
print((lambda x: x**2 + 5*x + 4) (-4))
```

**map**

```python
def add_five(x):
    return x + 5

nums = [11, 22, 33, 44, 55]
result = list(map(add_five, nums))
print(result)

print(list(map(lambda x: x+5, nums)))
```

**filter**

```python
nums = [11, 22, 33, 44, 55]
res = list(filter(lambda x: x%2==0, nums))
print(res)
```

**Generators**

```python
def countdown():
    i=5
    while i > 0:
        yield i
        i -= 1

for i in countdown():
    print(i)
```

**List Generators**

```python
def numbers(x):
    for i in range(x):
        if i % 2 == 0:
            yield i

print(list(numbers(11)))
```

**Decorators**

```python
def decor(func):
    def wrap():
        print("============")
        func()
        print("============")
    return wrap

@decor
def print_text():
    print("Hello world!")

print_text()
```

**Recursion**

```python
def factorial(x):
    if x == 1:
        return 1
    else: 
        return x * factorial(x-1)

print(factorial(5))
```

```python
def fib(x):
  if x == 0 or x == 1:
    return 1
  else: 
    return fib(x-1) + fib(x-2)
print(fib(4))
```

***args**

```python
def function(named_arg, *args):
    print(named_arg)
    print(args)

function(1, 2, 3, 4, 5)
```

****kwargs**

```python
def my_func(x, y=7, *args, **kwargs):
  print(x, y)
  print(args)
  print(kwargs)

my_func(2, 3, 4, 5, 6, a=7, b=8)
```

```python
def power(x, y):
  if y == 0:
    return 1
  else:
    return x * power(x, y-1)
		
print(power(2, 3))
```
