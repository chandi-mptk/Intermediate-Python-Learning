
**Working with Files**

**Opening Files**

```python
# write mode
open("filename.txt", "w")
```

```python
# read mode
open("filename.txt", "r")
open("filename.txt")
```

```python
# binary write mode
open("filename.txt", "wb")
```

```python
file = open("filename.txt", "w")
# do stuff to the file
file.close()
```

**Reading Files**


```python
data = '''Harry Potter
The Hunger Games
Pride and Prejudice
Gone with the Wind'''
file = open('books.txt', 'w')
file.write(data)
file.close()
```

```python
file = open('books.txt')
cont = file.read()
print(cont)
file.close()
```

```python
file = open('books.txt')
print(file.read(5))
print(file.read(7))
print(file.read())
file.close()
```

```python
file = open("books.txt")

for line in file.readlines():
    print(line)
    
file.close()
```

```python
file = open("books.txt")

for line in file:
    print(line)
    
file.close()
```

**Writing Files**

```python
file = open("newfile.txt", "w")
file.write("This has been written to a file")
file.close()

file = open("newfile.txt", "r")
print(file.read())
file.close()
```

```python
file = open("books.txt", "a")

file.write("\nThe Da Vinci Code")
file.close()

file = open("books.txt", "r")
print(file.read())
file.close()
```

```python
msg = "Hello world!"
file = open("newfile.txt", "w")
amount_written = file.write(msg)
print(amount_written)
file.close()
```

**With**

```python
with open("books.txt") as f:
  print(f.read())
```
