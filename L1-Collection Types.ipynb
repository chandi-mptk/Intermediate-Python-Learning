**Dictionaries**

```python
ages = {"Dave": 24, "Mary": 42, "John": 58}
print(ages["Dave"])
print(ages["Mary"])
```

```python
nums = {
    1: "one",
    2: "two",
    3: "three",
}
print(1 in nums)
print("three" in nums)
print(4 not in nums)
```

```python
pairs = {1: "apple",
  "orange": [2, 3, 4], 
  True: False, 
  12: "True",
}

print(pairs.get("orange"))
print(pairs.get(7, 42))
print(pairs.get(12345, "not found"))
```

**Sets**

```python
num_set = {1, 2, 3, 4, 5}

print(3 in num_set)
```

```python
nums = {1, 2, 1, 3, 1, 4, 5, 6}
print(nums)
nums.add(-7)
nums.remove(3)
print(nums)
nums.add(1)
print(nums)
```

**Sets can be combined using mathematical operations.**

```python
first = {1, 2, 3, 4, 5, 6}
second = {4, 5, 6, 7, 8, 9}

print(first | second)
print(first & second)
print(first - second)
print(second - first)
print(first ^ second)
```

**List Comprehensions**

```python
cubes = [i**3 for i in range(5)]

print(cubes)
```

```python
evens=[i**2 for i in range(10) if i**2 % 2 == 0]

print(evens)
```

```python
text = input()
dict = {}
for l in text:
    if l in dict:
        dict[l] += 1
    else:
        dict[l] = 1
print(dict)
```
